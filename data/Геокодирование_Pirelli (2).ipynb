```python
import pandas as pd
from geocoder import geocode_df
```

```python
!wget https://github.com/polina-jpg/pirelli_rs/raw/refs/heads/main/data/data_pirelli_geocode.xlsx
```

```python
input_file = "data_pirelli_geocode.xlsx"

plants = pd.read_excel(input_file, sheet_name="plants")
warehouses = pd.read_excel(input_file, sheet_name="warehouses")
cities = pd.read_excel(input_file, sheet_name="cities")

print("plants:", plants.shape)
print("warehouses:", warehouses.shape)
print("cities:", cities.shape)
```

```python
plants_locations = plants[["id", "city", "country"]].copy()
plants_locations["type"] = "plant"

warehouses_locations = warehouses[["id", "city", "country"]].copy()
warehouses_locations["type"] = "warehouse"

cities_locations = cities[["id", "city", "country"]].copy()
cities_locations["type"] = "city"

locations = pd.concat(
    [
        plants_locations,
        warehouses_locations,
        cities_locations,
    ],
    ignore_index=True
)

locations = locations[["type", "id", "city", "country"]]

print(f"Число строк: {len(locations)}")
locations.head()
```

```python
result = geocode_df(
    locations,
    address=None,
    city="city",
    region=None,
    country="country",
    user_agent="pirelli_emml_geocode",
    min_delay_seconds=1.1,
    max_retries=3,
    language="en",
)
```

```python
result
```

```python
output_file = "pirelli_data_coor.xlsx"

result.to_excel(output_file, index=False)

print(f"\nСохранено в {output_file}")
```
